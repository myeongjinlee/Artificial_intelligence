# Linear Regression (Boston dataset)

## Import Module

```python
import numpy as np
import requests
import pandas as pd
```

## Load Dataset

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# X = features, y = label
X = boston['data']
y = boston['target']
```

```python
# Make DataFrame
data = pd.DataFrame(X, columns = boston['feature_names'])
data['target'] = y
```

## Print Dataset

```python
# data
data.head()
```

## Gradient Descent

### Gradient Descent : one method to fit the linear model

```python
num_epoch = 100000 #???
learning_rate = 0.000000003
```

```python
# Parameter 
w = np.random.uniform(low=0.0, high=1.0, size=X.shape[1])
b = np.random.uniform(low=0.0, high=1.0)
```

```python
for epoch in range(num_epoch):
    # Predict y (=target)
    y_predict = X.dot(w) + b
    
    # Calculating Error
    error = np.abs(y_predict - y).mean()
    if error < 1:
        break
    
    # Gradient Descent for Multiple Variables
    w = w - learning_rate * ((y_predict - y)).dot(X)
    b = b - learning_rate * (y_predict - y).mean()
    
    # Print Error
    if epoch % 10 == 0:
        print("{0:6} error = {1:.5f}".format(epoch, error))
```

```python
print("----" * 10)
print("{0:5} error = {1:.5f}".format(epoch, error))
```

## Predict

```python
# Predict using calculated parameter
predict = X.dot(w) + b
```

```python
# Make Result DataFrame
result = pd.DataFrame({'actural': y, 'predict': predict})
result['difference'] = np.abs(y - predict)
```
